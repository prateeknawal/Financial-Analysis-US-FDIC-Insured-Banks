#### ENCODING ALL THE CATEGORICAL COLUMNS WITH THE HELP OF TARGET ENCODING AND ONE HOT ENCODING BELOW

1) Detecting categorical columns and trying to encode them

2) Using target encoding and hot encoding on the categorical columns

3) If the category has less than 10 classes then One hot encoding is used otherwise target encoding is done

```python
import os
from sklearn.preprocessing import OneHotEncoder
from category_encoders import TargetEncoder
import pandas as pd
from copy import deepcopy
import dill

# Columns to drop from ML models
cols_to_drop = []
# Categorical encoders dictionary
cat_encoders = {}
# New categorical (encoded) columns
cat_enc_columns = []

# Helper function for concatenating results
def apply_encoding(enc, X, columns, index):
    result = enc.transform(X)
    return pd.DataFrame(result, columns=columns, index=index)

# Encoding loop
for col in X_train.select_dtypes(include=['object']).columns:
    if X_train[col].nunique() < 10:  # One-hot encoding for low-cardinality features
        print("One-hot encoding of ", col)
        enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
        enc.fit(X_train[[col]])
        ohe_columns = [f"{col}_{x}" for x in enc.categories_[0]]
        
        # Apply one-hot encoding to train and test
        X_train[ohe_columns] = apply_encoding(enc, X_train[[col]], ohe_columns, X_train.index)
        X_test[ohe_columns] = apply_encoding(enc, X_test[[col]], ohe_columns, X_test.index)
        
        # Store the encoder
        cat_encoders[col] = [deepcopy(enc), "ohe"]
        cat_enc_columns.extend(ohe_columns)
    else:  # Target encoding for high-cardinality features
        print("Target Encoding of ", col)
        enc = TargetEncoder()
        enc.fit(X_train[col], y=y_train)
        new_col_name = f"{col}_trg"
        
        # Apply target encoding to train and test
        X_train[new_col_name] = enc.transform(X_train[[col]])
        X_test[new_col_name] = enc.transform(X_test[[col]])
        
        # Store the encoder
        cat_encoders[col] = [deepcopy(enc), "trg"]
        cat_enc_columns.append(new_col_name)
    
    # Track columns to drop
    cols_to_drop.append(col)

# Drop original categorical columns
X_train.drop(cols_to_drop, axis=1, inplace=True)
X_test.drop(cols_to_drop, axis=1, inplace=True)
```
