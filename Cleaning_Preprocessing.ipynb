```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"Data\FDIC_Insured_Banks.csv")
df
```

```python
#check for missing values
df.isnull().sum()
```

```python
# Drop rows where 'STATUS' and 'SCORE' columns have missing values
df = df.dropna(subset=['STATUS', 'SCORE'])

#Verify for missing values removed
df.isnull().sum()
```

```python
#Drop cells with 0 Deposits
df = df[df['DEPSUMBR'] != 0]
df
```

```python
df.dtypes
```

```python
#Convert "REPDTE" to Date column
df['REPDTE'] = pd.to_datetime(df['REPDTE'])

# Convert to the desired format 'YYYY/MM/DD'
df['REPDTE'] = df['REPDTE'].dt.strftime('%Y/%m/%d')
df['REPDTE']
```

```python
#Drop irrelevant columns
drop_irrelevant_columns = ['x', 'y', 'GeocodeSource', 'BKMO']
df = df.drop(columns = drop_irrelevant_columns)
df
```

```python
#Replace <None> with a specific value like 'Unknown'
df['CBSANAMB'] = df['CBSANAMB'].replace('<None>', 'Unknown')
df
```

```python
Cleaned_FDIC_Insured_Banks_path = 'Data\Cleaned_FDIC_Insured_Banks.csv'

df.to_csv(Cleaned_FDIC_Insured_Banks_path, index=False)
```
